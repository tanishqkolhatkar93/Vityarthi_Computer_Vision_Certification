```python
import cv2
import numpy as np
from PIL import Image
import requests
from io import BytesIO
from matplotlib import pyplot as plt
```

**Load image from web**

```python
response = requests.get('https://github.com/opencv/opencv/wiki/logo/OpenCV_logo_no_text.png')
image = Image.open(BytesIO(response.content))
```

**Convert image to OpenCV format**

```python
img = cv2.cvtColor(np.array(image), cv2.COLOR_RGB2BGR)
```

**Flip image horizontally**

```python
flipped_img = cv2.flip(img, 1)
```

**Convert the image to grayscale**

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

**Apply contrast detection using adaptive thresholding**

```python
_, thresholded_img = cv2.threshold(gray_img, 127, 255, cv2.THRESH_BINARY)
```

**Display the original, flipped, and thresholded images**

```python
plt.figure(figsize=(10, 6))

plt.subplot(131)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

plt.subplot(132)
plt.title('Flipped Image')
plt.imshow(cv2.cvtColor(flipped_img, cv2.COLOR_BGR2RGB))

plt.subplot(133)
plt.title('Thresholded Image')
plt.imshow(thresholded_img, cmap='gray')

plt.tight_layout()
plt.show()
```
