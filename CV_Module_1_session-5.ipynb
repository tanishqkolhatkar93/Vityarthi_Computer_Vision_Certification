```python
import numpy as np
import matplotlib.pyplot as plt
```

**Reading Image**

```python
import cv2
image = cv2.imread("baby.jpg")
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.title("Original")
plt.imshow(image_rgb)
```

**Saving Grayscale Image**

```python
gray_image = cv2.imread("baby.jpg", cv2.COLOR_BGR2GRAY)

cv2.imwrite("baby_gray.jpg", gray_image)
plt.title("Gray-Scale")
plt.imshow(gray_image)
```

**Loading and displaying saved image**

```python
#Saving
img = cv2.imread("baby_gray.jpg")

#Displaying
plt.title("Saved Image")
plt.imshow(img)
```
